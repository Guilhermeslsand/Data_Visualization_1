```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
brasil.head(5)
```

```python
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil.head(5)
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990', '1995', '2000', '2005', '2010'])
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.title('Imigrantes do Brasil ao Canadá')
plt.show()
```

Brasil X Argentina

```python
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano':argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina.head(5)
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
plt.plot(dados_argentina['ano'], dados_argentina['imigrantes'], label='Argentina')
plt.xticks(['1980','1985','1990', '1995', '2000', '2005', '2010'])
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.title('Imigrantes do Brasil ao Canadá')
plt.legend()
plt.show()
```

```python
#fig é para criar um figura, um espaço em branco
#ax é para os eixos da figura
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
ax.set_title('Imigrantes do Brasil ao Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
# A frequencia dos ticks do eixo x ser 5
ax.xaxis.set_major_locator(plt.MultipleLocator(3))
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,6))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
axs[0].set_title('Imigrantes do Brasil ao Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
# A frequencia dos ticks do eixo x ser 5
axs[0].xaxis.set_major_locator(plt.MultipleLocator(3))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot dos imigrantes do Brasil ao Canadá\n1980 a 2013')
axs[1].set_ylabel('Número de imigrantes')
axs[1].set_ylabel('Brasil')
axs[1].grid()

plt.show()
```

<h2>Comparação Países Sul America</h2>

```python
fig, axs = plt.subplots(4,2, figsize=(20,15))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013',
             fontsize=24)

paises = ['Brasil', 'Argentina', 'Colômbia', 'Peru']

ymin = 0
ymax = 7000

for i, pais in enumerate(paises):

  df_pais = pd.DataFrame(
      {'ano':df.loc[pais, anos].index.tolist(),
       'imigrantes': df.loc[pais, anos].values.tolist()})

  axs[i,0].plot(df_pais['ano'], df_pais['imigrantes'], lw=3, label=pais)
  axs[i,0].set_title(pais,fontsize=18,loc='left')
  axs[i,0].set_xlabel('Ano',fontsize=16)
  axs[i,0].set_ylabel('Número de imigrantes',fontsize=16)
  # A frequencia dos ticks do eixo x ser 3
  axs[i,0].xaxis.set_major_locator(plt.MultipleLocator(3))
  axs[i,0].xaxis.set_tick_params(labelsize=12)
  axs[i,0].yaxis.set_tick_params(labelsize=12)
  axs[i,0].set_ylim(ymin, ymax)
  axs[i,0].grid(linestyle='--')

  axs[i,1].boxplot(df_pais['imigrantes'])
  axs[i,1].set_title(f'Boxplot {pais}',fontsize=18)
  axs[i,1].set_ylabel('Número de imigrantes',fontsize=16)
  axs[i,1].grid(linestyle='--')

# for ax in axs.flat:
#      ax.xaxis.set_major_locator(plt.MultipleLocator(3))

# for ax in axs.ravel():
#     ax.set_ylim(ymin, ymax)

plt.show()
```

```python
cores = []

america_sul = df.query("Região == 'América do Sul'").sort_values('Total', ascending=True)

for pais in america_sul.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Brasil foi o 4º país com  mais imigrantes \npara o Canáda entre 1980 a 2013', fontsize = 18)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


for i, v in enumerate(america_sul['Total']):
  #ha -> alinhamento horizontal aliagend
  #va -> alinhamento vertical aliagend
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

#Tirar o frame -> as linhas
ax.set_frame_on(False)
#Tirar os valores do eixo
ax.get_xaxis().set_visible(False)
# Remover os ticks
ax.tick_params(axis='both', which='both', length=0)


plt.show()
```
